```python
import pandas as pd
```

```python
##CARGAR LOS ARCHIVOS
juegos= pd.read_csv("Analítica de datos en las organizaciones/VGsales.csv")
juegos
```

```python
##CONTAR POR GRUPOS CUANTOS JUEGOS SON DE PLATAFORMA Y GENERO UNICAMENTE LOS DE ACCION Y AVENTURA

juegosAyA= juegos[((juegos["Genre"]=="Adventure") | (juegos["Genre"]=="Action"))]
juegosAyA
```

```python
juegosAyA[["Plataforma","Rank","Genre"]].groupby(["Plataforma","Genre"]).count()
```

```python
##CONTAR POR GRUPOS CUANTOS JUEGOS SON POR GENERO Y ANIO UNICAMENTE DE LOS DE LOS ANIOS DEL 2010 Y SUPERIORES 
```

```python
juegos2010= juegos[(juegos["Year"]>=2010.0)]
juegos2010
```

```python
juegos2010[["Genre","Year","Rank"]].groupby(["Genre","Year"]).count()
```

```python
##USAR LA FUNCION SUM SOBRE LAS VENTAS GLOBALES

##AGRUPAR Y SUMAR LAS VENTAS GLOBALES POR GENERO

juegos[["Genre","Global_Sales"]].groupby(["Genre"])["Global_Sales"].sum()
```
